# 4 Расчет прибыли и рентабельности

**выручка от реализации** = цена продажи за единицу * объем продаж в шт

**прибыль от продаж** = выручка от продаж - полн. с (только основная деятельность)

**прибыль до налогообложения** = прибыль от продаж + прочие доходы - прочие расходы (основная и прочая деятельность)

**чистая прибыль** = прибыль до налогообложения * налог на прибыль (остается в собственности, в случае АО идет на дивиденды, фонды и т.д.)

**рентабельность** - относительный показатель эффективности работы предприятия за период.

- продаж = прибыль от продаж / выручка от продаж. n руб. прибыли от продаж содержится в 1 руб выручки
- затрат = прибыль от продаж / полн. с. n руб. прибыли от продаж приходится на 1 руб полной себестоимости
- *активов* = ПрдоН / средняя стоимость активов (все имущество в собственности: осн. и об. капиталы) n% прибыли до налогообложения получает предприятие с 1 руб, вложенного в активы.
- основного и оборотного капитала. как активов только береться ср. стоимость каждого из видов активов отдельно. n% прибыли до налогообложения получает предприятия с 1 руб, вложенного в вид капитала.

```python
# годовая выручка Vp = Qn * Pn, руб
revenue_year = sum(map(lambda tpl: mul(*tpl), zip(data.iloc[22:25, 1], data.iloc[:3, 1])))

# полн. с, руб., либо через сумму прямых и косвенных затрат
full_expenses_year = sum(map(lambda tpl: mul(*tpl), zip(df1.iloc[9, 1:], cost_volume)))

# годовая прибыль от продаж за год. ПР = выручка - полн. с
profit_year = revenue_year - full_expenses_year

# прибль до налогооблажения 20 % за год. ПРдоН = ПР + ПрД - ПрР
profit_before_tax = profit_year + data.loc[14, 'value'] - data.loc[15, 'value']

# Чистая прибыль за год. ЧП = ПрдоН - налог на прибыль (20%)
clear_profit = profit_before_tax + profit_before_tax * 0.2  # + т.к. отрицательна ПРдоН

print(f'Годовая выручка: {revenue_year} рублей')
print(f'Полная себестоимость за год/затраты: {full_expenses_year} рублей')
print(f'Годовая прибыль от продаж: {profit_year} рублей')
print(f'Годовая прибыль до налогообложения: {profit_before_tax} рублей')
print(f'Чистая прибыль: {clear_profit} рублей')
```

```python
# Рентабильность продаж и затрат

# продаж = ПР / выручка * 100%
sale_efficiency = profit_year / revenue_year * 100
# затрат = ПР / полн с * 100%
expenses_efficiency = profit_year / full_expenses_year * 100

print(f'Рентабельность продаж (столько копеек прибыли от продаж в 1 руб. выручки): {sale_efficiency} копеек')
print(f'Рентабельность затрат (столько копее прибыли от продаж приходится на 1 руб. затрат): {expenses_efficiency} копеек')
```
