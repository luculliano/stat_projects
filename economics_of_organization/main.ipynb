# 1 Расчет амортизационных отчислений

**Основной капитал** - материальные активы/имущество предприятия, многократно участвующие в производственном процессе и постепенно переносящие св. стоимость на себестоимость.

*Состоит из*:

- осн. средства. склады, оборудование, тс, инвентарь.
- нематериальные активы. патенты, авторские права.
- финансовые. вложения сроком более 12 мес. инвестиции в виде облигаций.

**Структура основного капитала** - доля каждой составляющей основного капитала в общей совокупности на дату.

**Стоимость основных средств**:

- первоначальная. Вкл. все расходы на приобретение, изготовление и довоедния до эксплуатации.
- остаточная. Учитывают износ. = Sперв. - Sa, где Sa - сумма амортизации.
- ликвидационная. Стоимость в момент списания.
- среднегодовая. Учитывают выбытие(списание) и ввод в эксплуатацию в течение года.
- восстановительная. Делают переоценку первоначальной стоимости в тек. условиях. По ней уже оценивают состояние основных средств, т.к. величина показательна.

**Износ осн. средства** - уменьшение первоначальной стоимости осн. средства в результате деятельности (физический) или научно-технического прогресса (моральный).

---

**Амортизация** - постепенный перенос стоимости собственных *основных средств* на себестоимость продукции с целью дальнейшего их восполнения.

*Способы начисления*:

- *линейная*. Равными долями в течение всего срока полезного использования.

- *уменьшенным остатоком*. Равномерно убывающими долями от макс. в начале срока ПИ до мин. к концу. 
Стоимость основного средства полностью не переносится на себестоимость

- *отдельно для каждоого года*. Равномерно убывающими долями. Отличается от уменьшенного остатка тем, что полностью переносит стоимость осн. средства на себестоимость.

- *пропорционально объему продукции*. Применяют в случае прямой зависимости срока службы осн. средства от частоты его эксплуатации

**Норма амортизации** - установленный процент отчислений в %.

**Сумма амотизации** - стоимостная оценка амортизации в абс. единицах.

Амортизацию относят к себестоимости (РСЭО), эта сумма идет в амортизационный фонд, который исп. для слудующего восполнения основных средств.

Амортизация учитывает износ + уменьшает налоговую базу + является источником восполнения осн. ср-в.

# 2 Расчет себестоимости единицы продукции

**Себестоимость** - стоимостная оценка затрат (ресурсов: осн. средства, трудовые ресурсы, топливо, материал) на производство и *реальзицию товаров*/услуг (продажу).

**Группиовки затрат**:

1. **Смета затрат за период**

- материальные. сырье и проч.
- зп и премии
- страховые взносы
- амотизационные отчисления
- прочие расходы

На ее основе определяют себестоимость единицы продукции (полн. с / объем продукции), если продуция однородная.

Определяют и **структуру затрат** - удельный вес каждой составляющей затрат в общем отношении за период.

Анализ структуры затрат необходим для снижения себестоимости.

2. **Калькуляция** - расчет себестоимости единицы продукции в денежном выражении.

На ее основе определяют себестоимость единицы продукции.

**Осн. статьи калькуляции** (производственное предприятие):

- цеховая себестоимость, т.е. затраты подразделения: 
  1. осн. зп осн. рабочих
  2. доп. зп осн. рабочих
  3. отчисления с зп осн. рабочих
  4. топливо
  5. расходы на содержание и эксплуатацию оборудования (РСЭО)
  6. осн. материалы
  7. вспомогат. материалы
  8. полуфабрикаты
  9. расходы на освоение производства
  10. общепроизводственные расходы
- производственная себестоимость: цеховая себестоимость + 
  11. общехоз. расходы
  12. брак
  13. проч. производственные расходы.
- *Полная себестоимость*: производственная себестоимость +

**Осн. статьи калькуляции** (предприятие сферы услуг):

- полуфабрикаты

**Осн. статьи калькуляции** (торгового предприятие):

- товары для перепродажи
- полуфабрикаты
- премии к зп
- накладные расходы

Виды затрат по способу отнесения на себестоимость:

1. прямые. связаны с производственным процессом или оказанием услуги, т.е. их можно напрямую отнести к опр. товару, т.к. напряму связаны с его созданием. (зп осн рабочих, полуфабрикаты)
2. косвенные. не относятся напрямую к производству продукции, но связаны с производствам. (зп менеджера, комуналка, налоги, реклама).

Алгоритм расчета косвенных затрат зависит от способа распределения. Для производственного предприятия способы распределения: пропорционально осн. зп осн. рабочих, прямым затратам и заводской себестоимости. Для торгововго предприятия: прямым затратам, доле выручки каждого товара в общ. объеме выручки. Для сферы услуг: также как и торгового.

Выбирается признак (относительно чего: осн. зп. осн. рабочих) и категория затрат (внепроизводственные затраты) и считается коэф. косвенных расходов.

```python
import numpy as np
plain_expenses = pd.DataFrame({
    'Статьи прямых затрат': ['Материальные затраты, руб.', 'Основная заработная плата основных рабочих, руб.', 'Страховые взносы (на основную заработную плату основных рабочих), руб.'],
    'Товар 1': [np.NaN]*3,
    'Товар 2': [np.NaN]*3,
    'Товар 3': [np.NaN]*3},
    columns = ['Статьи прямых затрат', 'Товар 1', 'Товар 2', 'Товар 3'])

# add initial data to dataframe
plain_expenses.iloc[0, 1:] = data.iloc[3:6].value
plain_expenses.iloc[1, 1:] = data.iloc[6:9].value

# подсчет страховых взносов
a = plain_expenses.iloc[1, 1] * 0.3
c = plain_expenses.iloc[1, 2] * 0.3
b = plain_expenses.iloc[1, 3] * 0.3
plain_expenses.iloc[2, 1:] = a, b, c

plain_expenses
```

```python
from operator import mul

indir_expenses = np.array([['Расходы на содержание и эксплуатацию оборудования (РСЭО)', np.NaN],
         ['Цеховые расходы', np.NaN],
         ['Заводские расходы', np.NaN],
         ['Внепроизводственные расходы', np.NaN],
         ['Итого косвенные затраты', np.NaN]])
indir_expenses = pd.DataFrame(indir_expenses, columns=('Статьи косвенных затрат', 'Сумма за год, руб.'))

# add initial data
indir_expenses.iloc[1:3, 1]  = data.iloc[11:13].value
indir_expenses.iloc[0, 1] = sa  # РСЭО = сумме амортизации
indir_expenses.iloc[3, 1] = data.value[13]  # внепроизв. расходы = расходы на рекламу
indir_expenses.loc[4, ['Сумма за год, руб.']] = indir_expenses.iloc[:4, 1].sum()

indir_expenses
```

зп осн. рабочих за год, сдельная (объем товара * себестоимость затрат на товар)

Аналогично для материальных затрат  и страховых

Полуаем общ. прямые затраты

```python
cost_volume = data.iloc[0:3].value
year_salary = sum(map(lambda tpl: mul(*tpl), zip(plain_expenses.iloc[1, 1:].values, cost_volume)))
mater_expenses = sum(map(lambda tpl: mul(*tpl), zip(plain_expenses.iloc[0, 1:].values, cost_volume)))
strah_expenses = sum(map(lambda tpl: mul(*tpl), zip(plain_expenses.iloc[2, 1:].values, cost_volume)))
total_plain_cost = year_salary + mater_expenses + strah_expenses
total_plain_cost
```

1. Расходы на содержание и эксплуатацию оборудования распределяются пропорционально основной заработной плате основных рабочих (сдельной заработной плате).
2. Цеховые и заводские расходы распределяются пропорционально прямым затратам.
3. Внепроизводственные расходы распределяются пропорционально заводской себестоимости.

Считаю коэффы

```python
# на 1 рубль основной зп осн. рабочих приходится 12 копеек расходов на СЭО
k1 = indir_expenses.iloc[0, 1] / year_salary
# на 1 рубль прямых затрат приходится 8 копеек цеховых затрат
k2_ceh = indir_expenses.iloc[1, 1] / total_plain_cost
# На 1 рубль прямых затрат приходится 37 копейки заводских затрат
k2_prod = indir_expenses.iloc[2, 1] / total_plain_cost
k1, k2_ceh, k2_prod
```

```python
# df1 с прямыми и косвенными видами затрат на кажый товар
part_1 =indir_expenses.iloc[:-2, 0]
part_2 = plain_expenses.iloc[:, 0]
part_3 = np.concatenate((part_1[:-1], ['Цеховая себестоимость'], [part_1[2]], ['Заводская себестоимость']))
df1 = pd.DataFrame({'Статья затрат, руб.': np.concatenate((part_2, part_3))})
for k, v in {i: [np.NaN]*df1.shape[0] for i in plain_expenses.columns[1:]}.items():
    df1[k] = v
df1.iloc[:3, 1:] = plain_expenses.iloc[:, 1:]
df1
```

```python
# for i in df1.columns[1:]:
#     df1[i][3] = round(k1 * df1[i][1], 2)
# df1
for good in df1.columns[1:]:
    df1.loc[3, [good]] = k1 * df1.loc[1, [good]]  # РСЭО = коэф * зп по каждому товару
    df1.loc[4, [good]] = k2_ceh * sum(plain_expenses[good])  # Цеховые расходы = коэф * сумма прямых затрат по товару
    df1.loc[5, [good]] = sum(df1.loc[:4, [good]][good])  # Цеховaя себестоимость = сумма 0-4
    df1.loc[6, [good]] = k2_prod * sum(plain_expenses[good])  # заводские расходы = коэф * сумма прямых затрат по товару
    df1.loc[7, [good]] = df1.loc[5, [good]] + df1.loc[6, [good]]  # заводская себестоимость = цеховая + заводские расходы
    df1[good] = df1[good].round(2)

df1
```

```python
# считаю общ. заводскую себестоимость и определяю коэф коммерческих расходов
prod_expenses = sum(map(lambda tpl: mul(*tpl), zip(df1.iloc[7, 1:].values, cost_volume)))
# на 1 рубль производственной себестоимости приходится 2 копеек коммерческих затрат
k_commerce = indir_expenses.iloc[3, 1] / prod_expenses
prod_expenses, k_commerce
```

```python
# заношу в массив сумму коммерческих расходов и полной себестоимсости на единицу продукцции (т.е. товара)
commerce_expenses = (round(df1.loc[7, [good]][good] * k_commerce, 2) for good in df1.columns[1:])
df1.loc[len(df1)] = ['Внепроизводственные расходы', *commerce_expenses]
total_expenses = (df1.loc[7, [good]][good] + df1.loc[8, [good]][good] 
                  for good in df1.columns[1:])
df1.loc[len(df1)] = ['Полная себестоимость единицы продукции (ATC)', *total_expenses]
df1
```

# 3 Расчет нормируемого оборотного капитала

**Оборотный капитал** - активы, однократно участвующее в производственном процессе, полностью преносит св. стоимость на себестоимость продукции, восполняется в конце производственного цикла.

*Состоит из*:

- запасы. всего: готовой продукцции, сырья, полуфабрикатов, незаверш. продукции.
- дебиторская задолженность (дз). то что тебе должны.
- деньги в кассе и на балансе
- финанс. вложения сроком менее 12 месяцев

**Структура оборотного капитала** - удельный вес каждой составляющей об. кап в общем отношении в относительных величинах на дату.

*Структура об. кап. расчитвыается после группировки на*:

- материальные активы. все что существует по факту, все выше перечисленное кроме дз.
- по риску. мин. риск; малый риск; ср. риск; высокий риск
- по ликвидности
  1. ликвидные. дс, фин. вложения.
  2. ср. ликвидные. дз.
  3. неликидные. запасы.
- по сфере обращения
  - касаемо производства (запас сырья и т.д.)
  - касаемо непроизводства (готовая продукция, дз, дс, отгруж. товар)
- по особенностям контроля
  - нормируемые. простые запасы
  - ненормируемые. дз, отгруженный товар, денежные средства
- по формированию
  - собственные
  - заемные
  
Для оценки платежеспособности предприятия исп. **коэффы ликвидности**:

- К абс. ликв.
- К промежут. ликв.
- К тек. ликв.

*Стадии об. капитала*: ДC -> Т -> П -> T' -> ДC' (производственное предприятие). ДС -> Т -> ДC' (торговое).

Показатели **оборачиваемости** (на сколько эффективно исп. об. ср-ва за период):

- Скорость оборота/коэф. оборачиваемости. число оборотов за период.
- Длительность оборота.
- Коэф. загрузки об. ср-в. сколько об. ср-в нужно для получения 1 руб. выручки.
- рентабельность. n% с ПРдоН получает предприятие с 1 руб, влоденного в об. капитал.

**Нормирование об. капитала** - определение мин. величины об. ср-в для работы, т.е. **нормы** (в относит. величинах) или **норматива** (в абс. величинах). Т.е. задача нацти нормируемый об. капитал, его мин. значение. $Sнобк = Sз + Sнзп + Sгп$

норматив об. средств в виде запасов (Sз) общ в руб

```python
# среднедневное потребление каждого вида сырья = потребность (объем * мат. затраты) / 365
qc1 = (data.loc[3, 'value'] * data.loc[0, 'value']) / 365
qc2 = (data.loc[4, 'value'] * data.loc[1, 'value']) / 365
qc3 = (data.loc[5, 'value'] * data.loc[2, 'value']) / 365

# норматив запаса в днях = тек. запас + страх. запас
t1_now = 0.5 * data.loc[25, 'value']  # тек. запас = 0.5 * интервал поставок или 0.5*(интервал+доп веремя на приемку)
t2_now = 0.5 * data.loc[26, 'value']
t3_now = 0.5 * data.loc[27, 'value']

t1_strah = t1_now * 0.2  # страх запас = тек. запас *  выбранный процент
t2_strah = t2_now * 0.2
t3_strah = t3_now * 0.2

# норматив запасов в днях, формула выше
t_nz_1 = t1_now + t1_strah
t_nz_2 = t2_now + t2_strah
t_nz_3 = t3_now + t3_strah

# норматив запасов общ. в руб
sz_1 = t_nz_1 * qc1
sz_2 = t_nz_2 * qc2
sz_3 = t_nz_3 * qc3

sz_common = sz_1 + sz_2 +sz_3
sz_common
```

норматив об. средств в нез. производстве общ в руб (Sнзп)

считают если период товарооборота больше 24 часов.

```python
# среднеднесной выпуск продукции по полной себестоимости в руб = (полн. с товара * объем) / 365
q1_seb = (df1.loc[9, 'Товар 1'] * data.loc[0, 'value']) / 365  
q2_seb = (df1.loc[9, 'Товар 2'] * data.loc[0, 'value']) / 365
q3_seb = (df1.loc[9, 'Товар 3'] * data.loc[0, 'value']) / 365

# коэф нараст. затрат, показывающий степень готовности изделия = 
# (общ среднеднев. потрбление сырья / среднеднев. выпуск продукции) как A + ((1 - A) / 2)
A1 = q1_seb / qc1
A2 = q2_seb / qc2
A3 = q3_seb / qc3
k1_nz = A1 + ((1 - A1) / 2)
k2_nz = A2 + ((1 - A2) / 2)
k3_nz = A3 + ((1 - A3) / 2)

# норматив об. ср-в в нез. производстве общ в руб = коэф нараст затрат * среднеднев. выпуск продукц + длит. цикла
s_nzp1 = q1_seb * k1_nz * data.loc[16, 'value']
s_nzp2 = q2_seb * k2_nz * data.loc[17, 'value']
s_nzp3 = q3_seb * k3_nz * data.loc[18, 'value']

s_nzp_common = s_nzp1 + s_nzp2 + s_nzp3
s_nzp_common
```

норматив об. ср-в гоотовой продукции Sгп в руб

```python
# норматив готов продукц = среднеднев выпуск по полн себ * длит хранения на складе
s_gp1 = q1_seb * data.loc[19, 'value']
s_gp2 = q2_seb * data.loc[20, 'value']
s_gp3 = q3_seb * data.loc[21, 'value']
s_gp_common = s_gp1 + s_gp2 + s_gp3
s_gp_common
```

Нормируемый об. капитал

```python
norm_ob_capital = sz_common + s_nzp_common + s_gp_common
norm_ob_capital
```

# 4 Расчет прибыли и рентабельности

**выручка от реализации** = цена продажи за единицу * объем продаж в шт

**прибыль от продаж** = выручка от продаж - полн. с (только основная деятельность)

**прибыль до налогообложения** = прибыль от продаж + прочие доходы - прочие расходы (основная и прочая деятельность)

**чистая прибыль** = прибыль до налогообложения * налог на прибыль (остается в собственности, в случае АО идет на дивиденды, фонды и т.д.)

**рентабельность** - относительный показатель эффективности работы предприятия за период.

- продаж = прибыль от продаж / выручка от продаж. n руб. прибыли от продаж содержится в 1 руб выручки
- затрат = прибыль от продаж / полн. с. n руб. прибыли от продаж приходится на 1 руб полной себестоимости
- *активов* = ПрдоН / средняя стоимость активов (все имущество в собственности: осн. и об. капиталы) n% прибыли до налогообложения получает предприятие с 1 руб, вложенного в активы.

# 6 Расчет точки безубыточности, порога рентабельности, запаса финансовой прочности


*Затраты в зависимости от объемов производства делят*:

- постоянные fixed cost. все в пределах установленной мощности не зависимо от объемов производства, мощность может измениться (новый цех появиться), тогда изменяются, но все равно считаются постоянными.
  - комуналка
  - оклад и страховые взносы на него
  - амортизация линейным способом
- непостоянные variable cost. зависят от объема производства.
  - сдельная зп и страховые наисления на сдельную зп
  - материальные затраты (полуфабрикаты, топливо)

*total cost* = fixed + variable

*average fixed costs* - удельные постоянные, т.е. на единицу продукции.

*average variable cost* - удельные переменные, т.е. на единицу продукции.

*average total cost* - удельные общ, т.е. общ затраты на единицу.

ATC = TC / Q, где Q - обем продукции. Аналогично для AVC и AFC, но со св. значением.

---

**Показатели бузубыточности**:

- точка безубыточности - значение объема производства и продаж, когда прибыль от продаж равна 0. В шт.

- порог рентабельности - стоимостная точка бузубыточности в руб.

- запас финансовой прочности - как далеко предприятие ушло от порога рентабельности. в руб. т.е. на эту величину предприятие может снизить выручку, попадая в порог рентабельности. Счиается в абс. и относит. величинах.

- операционный рычаг - показывает, во сколько раз изменится прибыль при изменении выручки на 1 %.

**Маржинальный доход/маржа** = выручка от продаж - переменные затраты. Сумма остается у компании после вычита затрат. 

На единицу продукции маржа = цена - avc

Коэф. маржи - доля маржи в общ. выручке = маржа / выручка

Деление затрат на постоянные и переменные
