# 6 Расчет точки безубыточности, порога рентабельности, запаса финансовой прочности

*Затраты в зависимости от объемов производства делят*:

- постоянные fixed cost. все в пределах установленной мощности не зависимо от объемов производства, мощность может измениться (новый цех появиться), тогда изменяются, но все равно считаются постоянными.
  - комуналка
  - оклад и страховые взносы на него
  - амортизация линейным способом
- непостоянные variable cost. зависят от объема производства.
  - сдельная зп и страховые наисления на сдельную зп
  - материальные затраты (полуфабрикаты, топливо)

*total cost* = fixed + variable

*average fixed costs* - удельные постоянные, т.е. на единицу продукции.

*average variable cost* - удельные переменные, т.е. на единицу продукции.

*average total cost* - удельные общ, т.е. общ затраты на единицу.

ATC = TC / Q, где Q - обем продукции. Аналогично для AVC и AFC, но со св. значением.

---

**Показатели бузубыточности**:

- точка безубыточности - значение объема производства и продаж, когда прибыль от продаж равна 0. В шт.

- порог рентабельности - стоимостная точка бузубыточности в руб.

- запас финансовой прочности - как далеко предприятие ушло от порога рентабельности. в руб. т.е. на эту величину предприятие может снизить выручку, попадая в порог рентабельности. Счиается в абс. и относит. величинах.

- операционный рычаг - показывает, во сколько раз изменится прибыль при изменении выручки на 1 %.

**Маржинальный доход/маржа** = выручка от продаж - переменные затраты. Сумма остается у компании после вычита затрат. 

На единицу продукции маржа = цена - avc

Коэф. маржи - доля маржи в общ. выручке = маржа / выручка

```python
# Деление затрат на постоянные и переменные

df5 = pd.DataFrame({
    'Сумма за год, руб.': (np.NaN for _ in range(7)),
    'Вид затрат': ('fixed', 'variable', 'fixed', 
                   'fixed', 'fixed', 'variable', 'variable')},
    index=('Амортизационные отчисления (РСЭО)',
            'Материальные затраты',
            'Цеховые расходы',
            'Заводские расходы',
            'Затраты на рекламу',
            'Сдельная заработная плата',
            'Страховые взносы на сдельную заработную плату'),
)
df5.index.name = 'Статьи затрат'
df5['Сумма за год, руб.'] = (sa, mater_expenses, 
                                 indir_expenses.iloc[1, 1], indir_expenses.iloc[2, 1],
                                 indir_expenses.iloc[3, 1], year_salary, strah_expenses)
df5['Сумма за год, руб.'] = df5['Сумма за год, руб.'].astype(float)
df5
```

```python
# совокуп. пост. затарты за год
fc = df5[df5['Вид затрат'] == 'fixed']['Сумма за год, руб.'].sum()
# совокуп. переменные затраты за год
vc = df5[df5['Вид затрат'] == 'variable']['Сумма за год, руб.'].sum()

# совокупные затраты предприятия total cost, равны полной себестоимости
tc = fc + vc
print(f'fixed cost: {fc} rub')
print(f'variable cost: {vc} rub')
print(f'total cost: {tc} rub')
```

В данном разделе курсовой работы необходимо рассчитать порог рентабельности, запас финансовой прочности в рублях и в процентах, операционный рычаг (все показатели считаются по предприятию в целом).

```python
# порог рентабельности = FC / коэф маржи, коэф маржи = маржа / выручка, маржа = выручка - VC
margin = revenue_year - vc
k_margin = margin / revenue_year
efficiency_frame = fc / k_margin

# запас финансовой прочности
betaf_absolute = revenue_year - efficiency_frame  # в руб
betaf_relative = betaf_absolute / revenue_year * 100 # в %

# операционный рычаг = маржа / прибыль от продаж
oper_leverage = margin / profit_year

print(f'Порог рентабельности (необходимая сумма выручки чтобы уйти в 0): {efficiency_frame} рублей')
print(f'Маржа (сумма остается у компаниии после вычита переменных затрат): {margin} рублей')
print(f'Маржинальность/коэф маржи (каждый выруенный рубль принесс рублей прибыли): {k_margin}')
print(f'Запас финансовой прочности (можно снизить выручку на сумму не уходя в минус): {betaf_absolute} рублей')
print(f'Запас финансовой прочности (можно снизить на этот процент выучку не уходя в минус): {betaf_relative}%')
print(f'Операционный рычаг (прибыль изменится в этот процент при изменении выручки на 1%): {oper_leverage}%')
```
