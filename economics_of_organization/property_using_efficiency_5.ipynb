- капиталоемкость
- рентабельность осн. капитала

```python
# остаточная стоимотсь осн. ср-в на конец года = перв. стоимость оборудования - сумма амортизации.
# т.е. стоимость осн. ср-в на конец года
left_main_assets = data.loc[9, 'value'] - sa

# cр. стоимость осн. ср-в = (НГ + КГ) / 2
average_main_assets = (data.loc[9, 'value'] + left_main_assets) * 0.5

# капиталоотдача = выручка или обхем продаж / ср. стоимость осн. ср-в
ko = revenue_year / average_main_assets

# капиталоемкость = 1 / капитолоотдача
kb = 1 / ko

# рентабельность осн. капитала = ПРдоН/ср. стоимость осн. капитала
main_assests_efficiency = profit_before_tax / average_main_assets * 100

# коэф. оборачиваемости об. ср-в (скорость оборота) = выручка / ср. стоимость об. капитала
kob = revenue_year / norm_ob_capital  # исп. нормир. об. капитал

# длительность одного оборота = 365 / коэф оборачиваемости
one_ob_time = 365 / kob

# коэф. загрузки об. ср-в = 1 / коэф оборачиваемости
kzagr = 1 / kob

# рентабельность об. капитала
ob_assets_efficiency = profit_before_tax / norm_ob_capital * 100

print(f'Средняя стоимость основного капитала: {average_main_assets} рублей')
print(f'Капиталоотдача (1 руб вложенный в осн. кап приносит выручки): {ko} рублей')
print(f'Капиталоемкоть (необходимо вложить чтобы 1 руб выручки получить): {kb} рублей')
print(f'Рентабельность осн. капитала (получаем ПРдоН с 1 руб, вложенного в осн. капитал): {main_assests_efficiency} рублей')
print(f'Коэф. оборачиваемости/скорость оборота (1 руб. вложенный в об. капитал приносит выручки): {kob} рублей')
print(f'Длительность 1 оборота: {one_ob_time} дней')
print(f'Коэф. загрузки (для получения 1 руб. выручки вклажываем в об. капитал): {kzagr} рублей')
print(f'Рентабельность об. капитала (получаем ПРдоН с 1 руб, вложенного в об. капитал): {ob_assets_efficiency} рублей')
```
