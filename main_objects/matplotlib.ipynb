# matploblib

\- библиотека для визуализации данных в двумерной или трехмерной графики. Поддерживает много 
форматов векторной и растровой графики.

*documentation*: https://matplotlib.org/stable/plot_types/index.html

В основном используется только пакет pyplot.

**График в целом (chart)** - графический способ отображения данных.

*График* - при помощи кривых линий. Имеет заголовок (именования признака с ед. измерения). Используется для динамических рядов распределения.

*Гистограмма* - столбики размером показателя единицы признака. Используются для измерения данных за опр. период или илюсстрирования сравнения объектов.

Есть гистограмма частот - Y-частота, X-диапазон. (Т.е. по сгруппированным данным). Формы бывают симметричными (от центра вправи и лево одинаково уменьшатся частота), ассиметричными (слева больше всего, вправо все меньше), остальное распредление.

*Круговая диаграмма* - удельные доли одного целого.

---

Axes (charts) находятся в обхекте рисунка (Figure). Внутри Figure можно строить 1 или несколько axes. Axe имеет оси заголовок, легенду и проч.

*plot* - от graph plot, т.е. построение графика.

#### plot для вывода графика

f plt.plot(x: collection, y: collection) - исп. если один график.

m ax.plot - на нужно ax посторить что-то.

```python
figure, ax = plt.subplots()
x = range(-3, 4)
y = (9, 4, 1, 0, 1, 4, 9)
ax.plot(x, y)
ax.plot(x, tuple(map(abs, x)))
```

```python
figure, axes = plt.subplots(2, 2)

x = range(-3, 4)
y = (9, 4, 1, 0, -1, -4, -9)

axes[0, 0].plot(x, y)
```

#### plot configuration

color/c='#9442CA' or 'red' or 

linestyle/ls='-' or '--' or '-.' or ':'

marker='o' or '\*' or '^'

alpha=0.2 (transparency from 0 to 1)

linewidth=2.5 (inches)

can write so as 3 arg: 'o--g'

```python
figure, ax = plt.subplots()
x = np.arange(-3, 4)
ax.plot(x, x + 1, color='#9442CA', marker='o')
ax.plot(x, x + 2, color='red', ls='--', alpha=0.4)
ax.plot(x, x + 3, 'o--b', linewidth=2, markersize=4)
```

#### axes limits


```python
x = np.arange(-3, 4)
y = np.array((5, 3, 1, 0, 1, 3, 5))

ax.plot(x, y)
ax.set_xlim(0, 5)
ax.set_ylim(-2, 6)
figure
```

#### lables

```python
figure = plt.figure()
figure.add_subplot()
ax = figure.axes[0]

x = np.arange(-3, 4)
y = np.array((4, 3, 1, 0, 1, 3, 4))

ax.set_title('The title number one', fontsize=20)
ax.set_xlabel('Axe X')
ax.set_ylabel('Axe Y')

ax.plot(x, y, label='y=x^2')

ax.legend(loc='best')  # reqired to place ax lable
```

```python
# plt.clf() - clear figure
# plt.cla() - clear axe

# own axes tickets
ax.set_yticks(np.arange(-2, 5, 0.5))
ax.set_xticks(np.arange(-3, 5, 0.5))

# remove borders
ax.spines.top.set_visible(False)

# grid to show and hide
ax.grid()

figure
```

```python
fig = plt.figure(dpi=100, figsize=(10, 20), tight_layout=True)
available = ['default'] + plt.style.available
x = np.arange(-2, 8, .1)
y = .1 * x ** 3 - x ** 2 + 3 * x + 2
for i, style in enumerate(available):
    with plt.style.context(style):
        ax = fig.add_subplot(10, 3, i + 1)
        ax.plot(x, y)
    ax.set_title(style)
```

```python
fig, ax = plt.subplots()
x = np.array((1, 2, 3, 4 ,5, 2, 3, 3, 3, 3))
ax.hist(x, bins=3, linewidth=0.5, edgecolor="white")
ax.set(xlim=(1, 6), xticks=np.arange(1, 6),
       ylim=(0, 10), yticks=np.linspace(0, 10, 6), ylabel='frequency')
```
