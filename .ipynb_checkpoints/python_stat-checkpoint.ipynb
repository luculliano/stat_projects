## jupyter

\- проект, разрабатывающий среды разврботки/блокноты, позволяющие хранить код, картинги, md комментарии вместе.

jupyternotebook - web надстройка над ipython.

jupytelab - усовершенствованный jupyternotebook. Есть еще electron десктопное.

jupyterhub - как gitlab, но для ноутбуков.

Есть в pypi (jupyterlab, notebook)

Исп. формат ipynb. Конфиг представляет собой json формат.

```shell
jupyterlab --notebook-dir=my/path/to/notebooks

jupyter notebook
```

#### keybindings 

**two modes avalable**: edit and command 

**types of cells**: md, code. row

```
1. Shift + Enter: Run the current cell and move to the next one.
2. Alt + Enter: Run the current cell and insert a new one below.
3. Ctrl + Enter: Run the current cell and keep the cursor in the same cell.
4. Esc: Enter Command mode.
5. Enter: Enter Edit mode.
6. A: Insert a new cell above the current one in Command mode.
7. B: Insert a new cell below the current one in Command mode.
8. D, D: Delete the current cell in Command mode.
9. Shift + M: Merge selected cells in Command mode.
10. Ctrl + S: Save the notebook.
11. M: Change the current cell type to Markdown in Command mode.
12. Y: Change the current cell type to Code in Command mode.
13. H: Show all keyboard shortcuts in Command mode.
```

Use with poetry

```shell
poetry add -D jupyter
poetry shell
poetry run jupyter notebook
```

Add config in /home/.jupyter

```
jupyter notebook --generate-config
```

**Kernel** - основа notebook, выполняющая код на разныз языках.

## theory

#### двумерные дынные

\- данные состоящие из двух **переменных/признаков**. Анализ проводится для нахождения взаимосвзи между ними.
*Одномерные* соответственно из одного и *многомерные* из нескольких.

Example:

| Расходы | Доходы |
|---------|--------|
| 100     | 500    |
| 250     | 750    |
| 400     | 900    |
| 300     | 600    |

| Рост | 170 | 175 | 168 | 182 |
|------|-----|-----|-----|-----|

---

#### панельные данные/данные временных рядов

\- многомерные данные, представлющие собой значения, как правило, собранные за разные промежутоки вермени для одних и тех же **изучаемых объектов** (Человеки, Компании). Людей можно назвать **выборкой/изучаемой совокупностью**.

| Человек | Год | Доход   | Возраст | Секс |
|---------|-----|--------|---------|------|
| 1       | 2013| 20 000 | 23      | Ф    |
| 1       | 2014| 25 000 | 24      | Ф    |
| 1       | 2015| 27 500 | 25      | Ф    |
| 2       | 2013| 35 000 | 27      | М    |
| 2       | 2014| 42 500 | 28      | М    |
| 2       | 2015| 50 000 | 29      | М    |

**выборка/выборочная совокупность/статистическая сводка** - объект исследования.

---

#### Стадии статистического исследования

- Сбор данных
- Группировка данных
- Расчет абсолютных и относительных величин
- Расчет средних величин
- Расчет показателей вариации
- Расчет показателей ряда динамики

---

#### Группировка

\- объединение единиц совокупности в группы, имеющие общ признак и общ размерность. Прецесс преобразования данных называется **сводкой**.

| группы по размеру | Частота |
|---------|--------|
| до 100     | 5    |
| от 100 до 250     | 7    |
| от 250 до 400     | 9    |
| от 400      | 6    |

Каждое значение/группа называется **рядом распределения**.
